```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Life Expectancy Data.csv')
df.head(3)
```

```python
display(df.head())
display(df.isnull().sum())
display(df.info())
```

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    if df[col].isnull().any():
        df[col].fillna(df[col].mean(), inplace=True)

correlation_matrix = df[numerical_cols].corr()
display(correlation_matrix)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Life expectancy '], kde=True)
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.show()
```

### Chart Description: Distribution of Life Expectancy

**Chart Type:** Histogram with KDE (Kernel Density Estimation) curve

**Variables:** Life Expectancy (years)

**What the Chart Shows:**
This histogram displays the frequency distribution of life expectancy values across all countries in the dataset. The overlaid KDE curve provides a smoothed representation of the distribution pattern.

**Major Trends and Patterns:**
- The distribution shows a roughly bimodal pattern with peaks around 65-70 years and 75-80 years
- Most countries cluster in the higher life expectancy range (70+ years), indicating generally improving global health
- There is a noticeable tail on the left side, showing some countries with significantly lower life expectancies (below 60 years)
- The distribution is slightly left-skewed, with more variation in lower life expectancies

**Key Insights:**
- The bimodal nature suggests two distinct groups: developing countries with lower life expectancies and developed nations with higher values
- The concentration of data in the 70-80 year range indicates that most countries have achieved reasonable healthcare standards
- The lower tail represents countries facing severe health challenges (conflict zones, poverty, disease burden)
- Overall, the distribution reflects global health inequality with a majority trending toward improved longevity

### Chart Description: Life Expectancy vs Adult Mortality

**Chart Type:** Scatterplot

**What the Chart Shows:**
This scatterplot visualizes the relationship between adult mortality rates and life expectancy across all countries. Each point represents a country-year observation.

**Major Trends and Patterns:**
- Strong negative correlation (-0.696): Higher adult mortality is associated with lower life expectancy

**Key Insights:**
- Adult mortality is the strongest negative predictor of life expectancy in the dataset
- Countries with adult mortality below 100 per 1000 generally have life expectancies above 70 years
- High adult mortality (>300) is devastating, reducing life expectancy by 20-30 years
- This metric captures working-age deaths, indicating the health burden during productive years
- The relationship suggests that reducing adult mortality is critical for improving overall life expectancy

# **COMPREHENSIVE SUMMARY: What the Complete Chart Set Reveals About Factors Affecting Life Expectancy**

## Overview

This notebook presents a comprehensive visual and statistical analysis of life expectancy data across 193 countries over a 15-year period (2000-2015). Through distribution plots, scatterplots, time series analysis, and correlation matrices, we have examined how various socioeconomic, health, and demographic factors influence how long people live.

---

## **Major Findings: The Primary Determinants of Life Expectancy**

### **1. Education: The Most Powerful Predictor (Correlation: 0.715)**

**Key Insight:** Education emerged as the single strongest positive predictor of life expectancy, surpassing even economic factors.

**Why Education Matters:**
- Educated populations make better health choices
- Education enables access to better employment and resources
- Literacy facilitates understanding of health information and preventive care
- Strong correlation with Income Composition (0.796) shows education drives overall human development

**Policy Implication:** Investing in education yields massive public health dividends—every additional year of schooling translates to measurable gains in population longevity.

---

### **2. Adult Mortality: The Strongest Negative Factor (Correlation: -0.696)**

**Key Insight:** Deaths during working age (15-60 years) are the most devastating predictor of reduced life expectancy.

**Why Adult Mortality Is Critical:**
- Captures health burden during economically productive years
- High adult mortality (>300 per 1000) reduces life expectancy by 20-30 years
- Reflects failures in healthcare systems, infectious disease control, and violence prevention
- Closely tied to HIV/AIDS prevalence (-0.523 correlation)

**Policy Implication:** Reducing adult mortality through disease prevention, trauma care, and violence reduction should be a top public health priority.

---

### **3. Income & Human Development (Correlation: 0.692)**

**Key Insight:** The Income Composition of Resources index (measuring overall human development) is the second-strongest positive predictor.

**The Development Advantage:**
- Higher human development scores combine education, income, and health
- Developed countries consistently maintain 10-12 year life expectancy advantage
- This gap has persisted over 15 years despite global improvements
- Reflects systemic differences in infrastructure, healthcare access, and social support

**Policy Implication:** Comprehensive development strategies addressing multiple factors simultaneously are more effective than single-issue interventions.

---

### **4. Nutrition & Body Composition (BMI Correlation: 0.559)**

**Key Insight:** Adequate nutrition, as reflected by BMI, is essential for longevity—but the relationship is complex.

**Nutrition's Role:**
- Very low BMI (<20) strongly predicts reduced life expectancy (malnutrition indicator)
- Optimal BMI range (40-60 in this dataset) associates with best outcomes
- Malnutrition indicators (thinness metrics) show strong negative correlations (-0.47)
- Food security is fundamental to population health

**Policy Implication:** Addressing malnutrition in poorest nations should be prioritized alongside education and healthcare access.

---

### **5. Disease Burden: HIV/AIDS (Correlation: -0.556)**

**Key Insight:** HIV/AIDS prevalence is the second-strongest negative predictor after adult mortality.

**The AIDS Impact:**
- Substantial correlation with adult mortality (0.523)
- Concentrated in specific geographic regions
- Represents both a healthcare and socioeconomic challenge
- Progress possible with treatment access and prevention programs

**Policy Implication:** Continued investment in HIV/AIDS treatment access and prevention is critical for affected regions.

---

### **6. Immunization Coverage (Polio: 0.462, Diphtheria: 0.475)**

**Key Insight:** Vaccination coverage serves as both a direct health intervention AND a proxy for overall healthcare system quality.

**Why Immunization Matters:**
- Achieving >90% coverage associated with 15-20 year gains in life expectancy
- Polio and Diphtheria coverage highly correlated (0.674), indicating comprehensive programs
- Countries with high coverage typically have strong public health infrastructure
- Cost-effective intervention with massive population impact

**Policy Implication:** Universal vaccination programs are among the highest-return public health investments.

---

### **7. Economic Factors: GDP (0.430) and Healthcare Spending (0.382)**

**Key Insight:** Economic prosperity matters, but with significant caveats about efficiency and diminishing returns.

**The Money Paradox:**
- **GDP shows diminishing returns:** Steep gains at low GDP levels, plateaus at high GDP
- **Spending efficiency matters more than amount:** Total health expenditure shows weak correlation (0.208)
- **Quality beats quantity:** Some low-spending countries achieve excellent outcomes through efficient systems
- **GDP and spending highly correlated (0.888):** Suggests wealth enables spending, but spending alone doesn't guarantee outcomes

**Policy Implication:** Focus on healthcare system efficiency, universal access, and preventive care rather than just increasing spending.

---

### **8. The Alcohol Paradox (Correlation: 0.392)**

**Key Insight:** Higher alcohol consumption correlates with HIGHER life expectancy—but this is NOT causal.

**Understanding the Paradox:**
- Reflects confounding by wealth: developed nations consume more alcohol AND have better healthcare
- Correlation with GDP (0.319) and Schooling (0.498) explains the pattern
- **Critical lesson in correlation vs. causation**
- Alcohol itself remains harmful; the correlation reflects underlying socioeconomic factors

**Policy Implication:** This underscores the importance of multivariate analysis and avoiding simplistic causal interpretations.

## **Synthesis: The Multifactorial Nature of Longevity**

### **No Single Silver Bullet:**
Life expectancy is determined by a complex interplay of factors:
1. Educational attainment (strongest predictor)
2. Healthcare system quality (reflected in immunization, mortality)
3. Economic development (enabling infrastructure)
4. Nutrition and food security
5. Disease burden management
6. Social and political stability

### **The Three Pillars of Longevity:**

1. **EDUCATION:** Empowers individuals, enables health literacy, drives development
2. **HEALTHCARE ACCESS:** Preventive care, immunization, and system efficiency matter most
3. **ECONOMIC STABILITY:** Provides resources but requires wise investment for health gains

### **Synergistic Effects:**
- Factors don't work in isolation—education enhances economic prospects, which enable better healthcare
- Income Composition and Schooling correlate at 0.796
- Comprehensive development strategies outperform single-factor interventions
- Breaking poverty cycles requires addressing multiple determinants simultaneously

## **Policy Recommendations Based on the Data**

### **High-Priority Interventions:**

1. **Expand Educational Access:**
   - Universal primary and secondary education
   - Focus on developing nations where gaps are largest
   - Education for girls and women (known multiplier effect)

2. **Strengthen Healthcare Systems:**
   - Achieve universal immunization coverage (>90%)
   - Build primary care infrastructure
   - Focus on preventive medicine and public health
   - Efficient spending beats high spending

3. **Reduce Adult Mortality:**
   - Improve working-age health services
   - Address HIV/AIDS, tuberculosis, and infectious diseases
   - Violence prevention and trauma care
   - Maternal and reproductive health services

4. **Address Malnutrition:**
   - Food security programs in lowest-BMI regions
   - Agricultural development and distribution systems
   - Nutritional education

5. **Target Persistent Inequality:**
   - Focus resources on developing nations
   - Address systematic infrastructure gaps
   - Share best practices from high-performing, moderate-spending countries

### **Avoid Single-Factor Thinking:**
- Simply increasing healthcare spending without system reform yields poor returns
- Economic growth alone doesn't guarantee health improvements
- Address root causes (education, systems, access) rather than symptoms
